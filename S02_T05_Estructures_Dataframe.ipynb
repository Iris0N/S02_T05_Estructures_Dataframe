```python
airlineData.dropna(inplace=True)
pd.DataFrame(airlineData.isnull().sum()).T.rename(index={0:'null values'})
```

Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...)

```python
airlineData["AvgSpeed"] = round(60*airlineData["Distance"]/airlineData["AirTime"]) 

airlineData["TotalDelay"] = airlineData["CarrierDelay"] + airlineData["WeatherDelay"] + airlineData["NASDelay"] + airlineData["SecurityDelay"] + airlineData["LateAircraftDelay"]

airlineData["CRSArrTime"] = airlineData["CRSArrTime"].astype(int)
airlineData["LateArrival"] =  np.where(airlineData["CRSArrTime"] < airlineData["ArrTime"], True, False)

airlineData["LateArrival1H"] = np.where(airlineData["TotalDelay"] >= 60 , True, False)

print("Dataframe after adding new columns:")
airlineData.head()
```

Taula de les aerolÃ­nies amb mÃ©s endarreriments acumulats

```python
airlineData["UniqueCarrier"].unique()
```

```python
airlineData["UniqueCarrier"].nunique()
```

```python
TotalDelayBYAirline= airlineData.groupby("UniqueCarrier")["TotalDelay"].sum()
TotalDelayBYAirline =TotalDelayBYAirline.sort_values(ascending=False)
TotalDelayBYAirline.head(5)
```
